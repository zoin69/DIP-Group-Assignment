```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load the style transfer model
style_transfer_model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')

# Step 2: Load and preprocess images
def load_and_preprocess_image(image_path):
    img = tf.io.read_file(image_path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.resize(img, (512, 512))
    img = img / 255.0  # Normalize to [0, 1]
    img = img[tf.newaxis, :]  # Add batch dimension
    return img

content_image = load_and_preprocess_image(r'E:\SLIIT\Year 3\Semester 2\Digital Image Processing\Image Tool\Training Data\Style transfer data\image to use.jpeg')
style_image = load_and_preprocess_image(r'E:\SLIIT\Year 3\Semester 2\Digital Image Processing\Image Tool\Training Data\Style transfer data\painting.jpg')


# Step 3: Apply style transfer
stylized_image = style_transfer_model(content_image, style_image)[0]

# Step 4: Display the results
def show_images(content_img, style_img, stylized_img):
    plt.figure(figsize=(15, 15))

    plt.subplot(1, 3, 1)
    plt.title("Content Image")
    plt.imshow(content_img[0])  # Remove batch dimension
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.title("Style Image")
    plt.imshow(style_img[0])  # Remove batch dimension
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.title("Stylized Image")
    plt.imshow(stylized_img[0])  # Remove batch dimension
    plt.axis("off")

    plt.show()

# Show the images
show_images(content_image, style_image, stylized_image)
```
